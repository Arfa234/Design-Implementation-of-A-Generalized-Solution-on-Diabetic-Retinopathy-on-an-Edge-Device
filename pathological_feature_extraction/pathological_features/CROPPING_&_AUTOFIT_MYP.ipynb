# **CROPPING & AUTO-FITTING IMAGES**

```python
from lesion_counting_myp.functional import autocrop_fundus, pad_given_margins, autofit_fundus_resolution, reverse_autofit, open_image
from lesion_counting_myp.utils.visualization import plot_image
from PIL import Image
```

```python
#FOR RDR TRAIN IMAGES

import os
from glob import glob

# Define the directory containing images
image_dir = '/content/drive/MyDrive/FYDP/Datasets/DR_subset/train/Rdr'

# Define the output directory for processed images
output_dir = '/content/drive/MyDrive/FYDP/Datasets/DR_subset/Processed_Images/RDR_Cropped'
os.makedirs(output_dir, exist_ok=True)  # Create the folder if it doesn't exist

# Get a list of all image file paths in the directory
image_paths = glob(os.path.join(image_dir, '*.jpg'))

# Loop through each image and apply the transformations
for example_path in image_paths:
    print(f"Processing image: {os.path.basename(example_path)}")

    # Load the image
    img = open_image(example_path)

    # Autocrop the fundus image
    print("Autocrop fundus (removal of black borders)")
    cropped_img, roi, margins = autocrop_fundus(img)

    # Reverse autocrop to pad with black borders
    print("Reverse autocrop (pad with black borders)")
    padded_img = pad_given_margins(cropped_img, margins)

    # Autofit: Autocrop, resize, and pad to square resolution (720x720)
    print("Autofit: Autocrop, resize and pad to square resolution (720x720)")
    autofit_fundus, params = autofit_fundus_resolution(img, 720)

    #Converting array into Images
    autofit_fundus_image = Image.fromarray(autofit_fundus.astype('uint8'))

    # Save the autofit image to the output folder
    output_path = os.path.join(output_dir, f'autofit_{os.path.basename(example_path)}')
    autofit_fundus_image.save(output_path)  # Save the image

    print(f"Saved processed image to: {output_path}\n")

print("Processing and saving complete for all images.")
```
